```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
seed = 0 
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
df = pd.read_csv('sonar.csv',header=None)
df
```

```python
dataset = df.values
X=dataset[:,0:60].astype('float')
Y_obj=dataset[:,60]
```

```python
e = LabelEncoder()
e.fit(Y_obj)
Y=e.transform(Y_obj)
```

```python
model=Sequential()
model.add(Dense(24,input_dim=60,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,Y,epochs=200,batch_size=5)
print("\n Accuracy: %.4f" %(model.evaluate(X,Y)[1]))
```

# 모델의 정확도가 100%로 나오지만 정말 100% 정확도 모델이 만들어진것이 아니다

## 과적합이 발생한것, 모델이 학습 데이터셋 안에서는 일정 수준이상 예측 정확도를 보이지만 새로운 데이터에 적용하면 잘 맞지 않는 현상을 말한다

## 과적합은 층이 너무 많거나 변수가 복잡해서 발생하기도하고 테스트셋과 학습셋이 중복될 떄 생기기도 한다

# 이를 방지하기위해 학습하는 데이터셋과 테스트할 데이터셋을 완전히 구분한 다음 학습과 동시에 테스트를 병행하며 진행하는 것이 한 방법이다
