```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=15 #글자크기
matplotlib.rcParams['axes.unicode_minus']=False #한글 폰트 사용시 마이너스글자 꺠짐방지
```

```python
plt.plot(df.index,df['출생아 수'])
plt.plot(df.index,df['합계 출산율'])
```

```python

fig,ax1=plt.subplots(figsize=(10,7))

ax1.plot(df.index,df['출생아 수'],color='#ff822d')

ax2=ax1.twinx() # x 축을 공유하는 쌍둥이 axis

ax2.plot(df.index,df['합계 출산율'],color='#ffd100')
                     
```

```python

fig,ax1=plt.subplots(figsize=(13,5))
ax1.set_ylabel('출생아 수 (천명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='#ff822d')

ax2=ax1.twinx() # x 축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100')           
```

```python

fig,ax1=plt.subplots(figsize=(13,5))
ax1.set_ylabel('출생아 수 (천명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='#ff822d')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx,val+12,val,ha='center')

ax2=ax1.twinx() # x 축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100')           
```

```python

fig,ax1=plt.subplots(figsize=(13,5))
ax1.set_ylabel('출생아 수 (천명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='#ff822d')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx,val+12,val,ha='center')

ax2=ax1.twinx() # x 축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100',marker='o',markersize=15,linewidth=5,markeredgecolor='w',markeredgewidth=3)           
```

```python
fig,ax1=plt.subplots(figsize=(13,5))
fig.suptitle('출생아 수 및 합계 출산율')
ax1.set_ylabel('출생아 수 (천명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'],color='#ff822d')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx,val+12,val,ha='center')

ax2=ax1.twinx() # x 축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'],color='#ffd100',marker='o',markersize=15,linewidth=5,markeredgecolor='w',markeredgewidth=3)      
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val + 0.07, val , ha='center')
```
