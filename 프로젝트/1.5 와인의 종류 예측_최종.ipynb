```python
from keras.models import Sequential
from keras.layers.core import Dense
from keras.callbacks import EarlyStopping,ModelCheckpoint

import os
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
np.random.seed(3)
tf.random.set_seed(3)
```

```python
df_pre = pd.read_csv('wine.csv',header=None)
df = df_pre.sample(frac=0.15)
```

```python
dataset = df.values
X=dataset[:,0:12]
Y=dataset[:,12]
```

```python
model=Sequential()
model.add(Dense(30,input_dim=12,activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
MODEL_DIR ='./model/'
if not os.path.exists(MODEL_DIR):
    os.mkdir(MODEL_DIR)
```

```python
modelpath="./model/{epoch:02d}-{val_loss:.4f}.hdf5"#모델 저장 폴더 만들기
```

```python
checkpointer = ModelCheckpoint(filepath =modelpath,monitor='val_loss',verbose=1,save_best_only =True)#모델 업데이트 및 저장
```

```python
early_stopping_callback = EarlyStopping(monitor = 'val_loss',patience = 100) # 학습 자동 중단 설정
```

```python
model.fit(X,Y,validation_split=0.2,epochs=2000,batch_size=500,callbacks=[early_stopping_callback,checkpointer])
```
